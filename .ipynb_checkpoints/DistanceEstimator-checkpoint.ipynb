## Modeling

```python
class DistanceEstimator(nn.Module):
    def __init__(self):
        super(DistanceEstimator, self).__init__()
        
        #Layer
        self.model = nn.Sequential(
            nn.Linear(3,8),
            nn.LeakyReLU(negative_slope=0.01),
            
            nn.Linear(8,16),
            nn.LeakyReLU(negative_slope=0.01),
            
            nn.Linear(16,32),
            nn.LeakyReLU(negative_slope=0.01),
            
            nn.Linear(32,64),
            nn.LeakyReLU(),
            
            nn.Linear(64,128),
            nn.ReLU(),
            
            nn.Linear(128,64),
            nn.ReLU(),
            
            nn.Linear(64,32),
            nn.ReLU(),
            
            nn.Linear(32,16),
            nn.ReLU(),
            
            nn.Linear(16,1)
        )

    def forward(self, x):
        out = self.model(x)
        return out
```

## Make  variable

```python
model = DistanceEstimator()
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer,
                                                       factor=0.1,
                                                       mode='min', # ìš°ë¦¬ëŠ” ë‚®ì•„ì§€ëŠ” ê°’ì„ ê¸°ëŒ€
                                                       verbose=True)

model.to(device)
```

```python
# train parameters
def count_parameter(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)
count_parameter(model) # 22017
```

## Make Train, Valid function

```python
def train(model, train_dataloader, idx_interval):
    model.train()
    
    mse_loss = 0
    rmse_loss = 0
    
    for idx, batch in enumerate(train_dataloader):
        optimizer.zero_grad()
        
        prediction = model(batch[0].to(device))
        loss = loss_fn(prediction, batch[1].to(device)).cpu()
        
        # Backpropagation
        loss.backward()
        optimizer.step()
        
        if idx % idx_interval == 0:
            print("Train Epoch: {} [{}/{}] \t Train Loss(MSE): {:.4f}".format(epoch, batch_sz*(idx+1), \
                                                                            len(train_dataloader)*batch_sz, \
                                                                            loss.item()))
        mse_loss += loss.item()
    
    mse_loss /= len(train_dataloader)
        
    return mse_loss
```

```python
def evaluate(model, valid_dataloader):
    model.eval()
    
    valid_loss = 0
    
    with torch.no_grad():
        for idx, batch in enumerate(valid_dataloader):
            predictions = model(batch[0].to(device))
            loss = loss_fn(predictions, batch[1].to(device)).cpu()
            valid_loss += loss.item()
            
    valid_loss /= len(valid_dataloader)
    
    return valid_loss
```

```python
# Function to save the model 
def saveModel(model): 
    path = "./weights/ODD_basic.pth" 
    torch.save(model.state_dict(), path) 
```

## Train and Validation

```python
Epoch = 100
best_rmse = 99999

for epoch in range(1,(Epoch+1)):
    train_mse = train(model, train_dataloader, 250)
    valid_mse = evaluate(model, valid_dataloader)
    
    train_rmse = np.sqrt(train_mse)
    valid_rmse = np.sqrt(valid_mse)
    print("[Epoch: {} \t Valid MSE: {:.4f} \t Valid RMSE: {:.4f}]".format(epoch, valid_mse, valid_rmse))
    
    scheduler.step(valid_rmse)

    # Save model
    if valid_rmse < best_rmse:
        path = "./weights/ODD_basic.pth" 
        torch.save(model.state_dict(), path) # ëª¨ë¸ì˜ ê°€ì¤‘ì¹˜ë§Œ ì €ìž¥ êµ¬ì¡°ëŠ” ì €ìž¥ x..?
```

from sklearn.metrics import mean_squared_error
def train_model(model, train_dataloader, valid_dataloader, loss_fn, lr=1e-5, batch_size=512, epochs=100, validate=False):

  
  # Convert model parameters and buffers to CPU or Cuda
  model.to(device)

  best_rmse = np.Inf
  print("Begin training...") 
  for epoch in range(1, epochs+1): 
    running_train_loss = 0.0 
    running_rmse = 0.0 
    running_vall_loss = 0.0 
    total = 0 

    for batch_ind, samples in enumerate(train_dataloader):
      x_train, y_train = samples
      optimizer.zero_grad()
      pred = model.forward(x_train)
      train_loss = loss_fn(pred, y_train)
      train_loss.backward()
      optimizer.step()
      running_train_loss += train_loss.item()

    train_loss_value = running_train_loss/len(train_dataloader)
    with torch.no_grad(): 
      model.eval() 
      for data in valid_dataloader: 
        inputs, outputs = data 
        predicted_outputs = model(inputs) 
        val_loss = loss_fn(predicted_outputs, outputs) 
      
        # The label with the highest value will be our prediction 
        running_vall_loss += val_loss.item()  
        total += outputs.size(0) 
        rmse = mean_squared_error(outputs, predicted_outputs)**0.5
        running_rmse += rmse

    # Calculate validation loss value 
    val_loss_value = running_vall_loss/len(valid_dataloader)  
    rmse = running_rmse / total

    if rmse < best_rmse:
      saveModel(model)
      best_rmse = rmse

    # Print the statistics of the epoch 
    print('Epoch {0}/{1} - loss: {2:.4f} / val_loss: {3:.4f} - RMSE: {4:.4f}'.format(epoch, epochs, train_loss_value, val_loss_value,rmse))
